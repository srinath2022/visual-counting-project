# Generate combination synthetic video

```python
# pick random number of flippings
# TO-DO : pick random speed
# TO-DO : make random perturbations to the individual frames
# pick random 1-flip video, speed it up, make random perturbations and add to resulting video
from IPython.display import HTML
import base64

def show_video(video_path):
  mp4 = open(video_path, 'rb').read()
  data_url = 'data:video/mp4;base64,' + base64.b64encode(mp4).decode()
  return HTML("""<video width=600 controls>
      <source src="%s" type="video/mp4"></video>
  """ % data_url)

def getFrames(video_path):
    frames = []
    cap = cv2.VideoCapture(video_path)
    while cap.isOpened():
        ret, frame = cap.read()
        if ret is False:
            break
        
        # img = Image.fromarray(frame)
        # frames.append(img)
        frames.append(frame)
    
    cap.release()
    return frames

def generate_synthetic_video(MIN_FLIP = 1, MAX_FLIP = 30):
  num_flips = random.randint(MIN_FLIP, MAX_FLIP)
  gen_frames = []
  for i in range(num_flips):
    # pick a random 1-flip video
    randpath = random.choice(glob.glob(saveDirectoryPath+vidName+'*.mp4'))
    frames   = getFrames(randpath)
    gen_frames.extend(frames)

  return gen_frames

def frames_to_video(frames, path, fps=30):
  size = frames[0].shape
  print(size)
  out = cv2.VideoWriter(path, cv2.VideoWriter_fourcc(*'XVID'), fps, (size[1], size[0]))
  for i in range(len(frames)):
      # frame = cv2.flip(frames[i],0)
      # out.write(frame)
      out.write(frames[i])
  out.release()
  
```

```python
vid = generate_synthetic_video()
savePath = generatedDirPath+"sample5.mp4"
# savePath = "/tmp/sample5.mp4"
frames_to_video(vid, savePath, 60)
```
